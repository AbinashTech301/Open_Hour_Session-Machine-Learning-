The describe() function on the Pandas
DataFrame lists 8 statistical properties of each attribute. They are:
1. Count.
2. Mean.
3. Standard Deviation.
4. Minimum Value.
5. 25th Percentile.
6. 50th Percentile (Median).
7. 75th Percentile.
8. Maximum Value

```python
from pandas import set_option
set_option('display.width', 90)
set_option('precision', 3)
description = data.describe()
print (description)
```

### UNIVARIATE PLOTS
#### Univariate plots provide one way to find out about those properties (and univariate descriptive statistics provide another). There are two basic kinds of univariate, or one-variable-at-a-time plots, Enumerative plots, or plots that show every observation, and Summary plots, that generalize the data into a simplified representation.
 1. Histograms.
 2. Density Plots.
 3. Box and Whisker Plots.

```python
from matplotlib import pyplot
data.hist()
pyplot.show()
```
